```python
from sklearn.ensemble import RandomForestClassifier 

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1) 

train = sp500.iloc[:-100]
test = sp500.iloc[-100:] 

predictors = ["Close", "Volume", "Open", "High", "Low"] 
model.fit(train[predictors], train["Target"]) 
```

```python
from sklearn.metrics import precision_score 

preds = model.predict(test[predictors]) 
```

```python
import pandas as pd

preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds) 
```

This Precision score tells me that when I said the stock was going to go up, I was correct 57% of the time. In this case, this tells me that when the model said the stock price was going to increase, it was correct 57% of the time.

----------- Building a Backtesting System ------------

```python
def predict(train, test, predictors, model): 
    model.fit(train[predictors], train["Target"]) 
    preds = model.predict(test[predictors]) 
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1) 
    return combined 
```

Start Value = when you backtest you want to have a certain amount of data to train your first model. Every trading year has about 250 days, so this says takes 10 years of data, and train your first model with 10 years of data.  

Step Value = means we will be training a model for about a year and then moving into the next year, and the next year. 

We are going to take the first 10 years of data and predict values for the 11th year, and then take the first 11 years and predict values for the 12th year and on and on.  

all_predictions = a list of data frames where each data frame is the predictions for a single year

```python
def backtest(data, model, predictors, start=2500, step=250): 
    all_predictions = [] 

    for i in range(start, data.shape[0], step): 
        train = data.iloc[0:i].copy() 
        test = data.iloc[i:(i+step)].copy() 
        predictions = predict(train, test, predictors, model) 
        all_predictions.append(predictions) 
    return pd.concat(all_predictions) 
```

```python
predictions = backtest(sp500, model, predictors) 
```

```python
predictions["Predictions"].value_counts() 
```

This means we predicted the market would go down on about 3788 days, and we predicted the market would go up around 2671 days 

```python
precision_score(predictions["Target"], predictions["Predictions"]) 
```
